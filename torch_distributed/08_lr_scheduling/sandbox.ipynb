```python
from lightning_train import MyLightningModule
from argparse import Namespace
import torch
from torch.utils.data import Dataset, DataLoader
from simlm_dataloader import RetrievalDataLoader
from transformers import AutoTokenizer, AutoModel
from lightning_train import JsonlDataset


args = Namespace(
    model_name_or_path="intfloat/simlm-base-msmarco",
    train_path="data/train.jsonl",
    output_dir="runs/$MY_RUN_NAME",
    epochs=3,
    temperature=0.02,
    val_path="data/validation.jsonl", 
    val_check_interval=200,
    q_max_len=32,
    p_max_len=144,
    batch_size=16,
    simlm_dataloader=True,
    train_n_passages=16,
)

```

```python
tokenizer = AutoTokenizer.from_pretrained(args.model_name_or_path)
train_loader = RetrievalDataLoader(args=args, tokenizer=tokenizer)
```
